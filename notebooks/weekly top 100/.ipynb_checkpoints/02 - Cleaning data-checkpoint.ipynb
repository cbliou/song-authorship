```python
import re
import pandas as pd
import numpy as np
import csv
import json
import os
```

```python
library = "C:\\Users\\liblabs-user\\Desktop\\song-authorship\\data"
laptop = "not yet"
desktop = "C:\\Users\\Sam\\Desktop\\song authorship\\data"
```

```python
os.chdir(desktop)
df = pd.read_csv(open(desktop + "\\Weekly_data.csv"))
```

```python
def isnull(x):
    """
    """
    return x != x
```

```python
def clean_lyric(lyric):
    
    repl = {"\[.+?\]": "", "\s{2,}": "\n", "lyrics": "", "\(": "", "\)": ""}
    
    for i in repl:
        lyric = re.sub(i, repl[i], lyric, flags = re.I | re.S)
    
    return lyric
```

```python
filler = [" and ", " featuring ", " & ", " x ", " / "]
repl = {"(\"|”).+?(\"|”)": "", "\s+": " ", "\s,": ","}
textcols = ["Artists", "Name", "Genre", "Writing Credits", "Lyrics"]
#textcols = ["Genre"]
for i in textcols:
    for j in range(len(df)):
        if not isnull(df.loc[j, i]):
            if i == "Artists":
                for k in filler:
                    df.loc[j, i] = re.sub(k, "", df.loc[j, i], flags = re.I | re.S)
            elif i == "Lyrics":
                df.loc[j, i] = clean_lyric(df.loc[j, i])
            else:
                for k in repl:
                    df.loc[j, i] = re.sub(k, repl[k], df.loc[j, i], flags = re.I | re.S)
                df.loc[j, i] = df.loc[j, i].strip()
```

## Add songwriters

```python
writers = json.load(open(desktop + "\\composers.json", "r"))
writers[1] = "Tobias Jesso Jr"
writers[5] = "Jacob Kasher"
writers[10] = "Pop and Oak"
writers[16] = "Mike Will Made It"
```

```python
composers = []
for i in range(len(df)):
    for j in writers:
        if not isnull(df.loc[i, "Writing Credits"]):
            if re.search(j, df.loc[i, "Writing Credits"], flags = re.I | re.S) is not None:
                composers.append(True)
                break
    else:
        composers.append(False)
```

```python
df["Songwriter"] = np.array(composers)
df.to_csv("Weekly_data_clean.csv", index = False)
```
